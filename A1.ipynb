```python
import pandas as pd
df = pd.read_csv("API_SG.GEN.PARL.ZS_DS2_en_csv_v2_99.csv", skiprows=4, delimiter=",")
df = df.drop(columns=['Unnamed: 68', 'Indicator Name', 'Indicator Code', '1960', '1961', '1962', '1963', '1964',
       '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973',
       '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982',
       '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996'])
df = df.dropna()
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')
average_row = df.iloc[:, 2:].mean()
average_row = pd.DataFrame([['Average', 'AVG'] + average_row.tolist()], columns=df.columns)
df = pd.concat([df, average_row], ignore_index=True)

print(df.columns)
print(df.head())
print(df.tail())
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go


df_melted = df.melt(id_vars=['Country Name', 'Country Code'],
                     var_name='Year',
                     value_name='Value')


df_melted['Year'] = df_melted['Year'].astype(int)


custom_colors = {country: 'grey' for country in df_melted['Country Name'].unique()}
custom_colors['Average'] = 'red'  # Highlight the "Average" line


fig = px.line(df_melted,
              x='Year',
              y='Value',
              color='Country Name',
              title='More Women Are in Parliament Around The World',
              labels={'Value': 'Proportion (%)'},
              markers=False,
              color_discrete_map=custom_colors)

last_year = df_melted['Year'].max()
average_last_value = df_melted[(df_melted['Country Name'] == 'Average') & (df_melted['Year'] == last_year)]['Value'].values[0]
fig.add_trace(go.Scatter(
    x=[last_year],
    y=[average_last_value],
    text=["Average"],
    mode="text",
    textposition="middle right",
    showlegend=False
))

fig.show()
```
